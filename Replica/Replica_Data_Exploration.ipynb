```python
import pandas as pd
import pandas_gbq
import google.auth
```

```python
credentials, project = google.auth.default()
```

```python
#Read in a sql query from a file and run it against a BigQuery table.
sql = open('Thursday_Q4_2022.sql', 'r').read()
```

```python
df = pandas_gbq.read_gbq(sql, project_id=project, credentials=credentials)
```

```python
#group by travel_purpose and sum the segment_distance_travelled
df = df.groupby('travel_purpose').agg({'segment_distance_traveled': 'sum'}).reset_index()
```
