```python
import pandas as pd
from arcgis.features import FeatureLayer
```

```python
#Download HIN data
# get spatially enabled dataframe
def download_sdf(service_url):
    feature_layer = FeatureLayer(service_url)
    query_result = feature_layer.query()
    # Convert the query result to a list of dictionaries
    sdf = query_result.sdf
    return sdf

HIN_Network = download_sdf('https://maps.trpa.org/server/rest/services/Transportation_Planning/MapServer/8')
```

```python
total_miles=HIN_Network['Shape.STLength()'].sum()
HIN_Network_ratio = pd.DataFrame(columns=['HIN_Network','total_length','percent_included'])
networks = ['ped_HIN_05','car_HIN_05','bike_HIN_05','ped_HIN_0','car_HIN_0','bike_HIN_0','ped_HIN_tenth','car_HIN_tenth','bike_HIN_tenth']
print(total_miles)
for network in networks:
    total = HIN_Network.loc[HIN_Network[network]==1,'Shape.STLength()'].sum()
    #total_rows = len(tree_permits_2022)
    #yes_rows = len(tree_permits_2022[tree_permits_2022[reason] == 'CHECKED'])
    percentage_covered = (total / total_miles) * 100
    HIN_Network_ratio = pd.concat([HIN_Network_ratio,pd.DataFrame({'HIN_Network':[network],'total_length':[total],'percent_included':[percentage_covered]})],ignore_index=True)
HIN_Network_ratio['Miles']=HIN_Network_ratio['total_length']/1609.34
```

```python
HIN_Network_ratio.to_excel('HIN_Network_Percent.xlsx')  
```
